数据整理：标准化

```python
def standard_x(x):
    ma = x.max()
    mi = x.min()
    a = []
    for i in range(len(x)):
        a.append((x[i]-mi)/(ma-mi))
    return a

plt.figure(figsize=(14,7))
for i in data.columns[1:-1]:
    plt.plot(data['日期'], standard_x(data[i]), '--')

plt.show()
```

数据整理：计算收益率

算术收益率：[P(t) - P(t-1)] / P(t-1)

```python
data['浦发银行'][1:] - data['浦发银行'].shift(1)[1:]/data['浦发银行'].shift(1)[1:]
```

几何收益率：log(P(t)/P(t-1))

```python
np.diff(np.log(data['浦发银行'].astype('float64')))
```

# 2. 资本资产定价模型

## 2.1 最小二乘法

```python
import statsmodels.api as sm

rmf = df['上证指数'] - df['no_risk']
rif = df['浦发银行'] - df['no_risk']

y = rif
x = rmf
X = sm.add_constant(x)
model = sm.OLS(y,X).fit()
model.summary()
```

拟合参数

```python
model.params
```

拟合t值

```python
model.tvalues
```

拟合p值

```python
model.pvalues
```

拟合R平方

```python
model.rsquared
```

预测结果

```python
pre = model.predict()

plt.figure(figsize=(10, 7))
plt.scatter(x,y,color='b')
plt.plot(x,pre,color='r')
```

```python
import seaborn as sns

plt.figure(figsize=(10,7))
sns.regplot(x=x,y=y, color='r', ci=75)
```

## 2.2 数据分析

```python
def f(dname):
    rmf = df['上证指数'] - df['no_risk']
    rif = df[dname] - df['no_risk']
    y = rif
    x = rmf
    X = sm.add_constant(x)
    model = sm.OLS(y,X).fit()
    model.summary()
    para = model.params
    t_value = model.tvalues
    p_values = model.pvalues
    r_squared = model.rsquared
    return para, t_value, p_values, r_squared
```

```python
f('浦发银行')
```

```python
res = pd.DataFrame()

for i in data.columns[1:-2]:
    res[i] = f(i)

res.index=['para', 't_value', 'p_value', 'r_squared']
res.T.sort_values('r_squared')
```
